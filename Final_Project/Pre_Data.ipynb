## 讀取雨量資料

```python
Chih_Ruei = pd.read_csv("data/雨量/池端01_19.csv", index_col=0)
```

```python
Chih_Ruei = Chih_Ruei.rename({'Rainfall_y': 'Rainfall'}, axis=1)
Chih_Ruei
```

```python
Chih_Ruei['Time'] = pd.to_datetime(Chih_Ruei['Time'])
Chih_Ruei['Year'] = Chih_Ruei['Time'].map(lambda x: x.year)
Chih_Ruei['Month'] = Chih_Ruei['Time'].map(lambda x: x.month)
Chih_Ruei
```

```python
import os
from pathlib import Path
for i in range(2001, 2020):
    '''設定目的地路徑'''
    path_with_index = Path(str(i)+"_Chih_Ruei_Rainfall.csv")
    destination_path = os.path.join("data_processed/Chih_Ruei", path_with_index)
    if os.path.isfile(destination_path):
        pass
    else:
        '''遮罩-依照月份篩選'''
        filter = (Chih_Ruei["Year"] == i)
        df_by_month = Chih_Ruei[filter]
        df_by_month.to_csv(destination_path, index = False)
```

```python
X_CR = Chih_Ruei[["Time"]]
Y_CR = Chih_Ruei[["Rainfall"]]

plt.plot(X_CR, Y_CR)
plt.show()
```
